```python
import pandas as pd
import scipy.stats
```

```python
survey_df = pd.read_csv("2016-FCC-New-Coders-Survey-Data.csv")
survey_df
```

```python
survey_df = survey_df[["EmploymentField", "EmploymentStatus", "Gender", "JobPref", "JobWherePref", "MaritalStatus", "Income"]]
survey_df
```

```python
survey_df.Gender.unique()
```

```python
survey_df = survey_df[survey_df["Gender"].isin(['male', 'female'])]
survey_df
```

Исследуем связи между Gender и JobPref с помощью таблиц сопряженности

```python
df_1 = survey_df[["Gender", "JobPref"]].dropna()
df_1
```

```python
df_1.Gender.unique()
```

```python
df_1.JobPref.unique()
```

```python
resualt = pd.crosstab(df_1.Gender, df_1.JobPref, margins=False)  # Таблица сопряженности с наблюдаемыми значениями 
resualt
```

```python
expected = scipy.stats.contingency.expected_freq(resualt)       # Таблица сопряженности с ожидаемыми значениями 
pd.DataFrame(expected, index=resualt.index, columns=resualt.columns) 
```

Значения в ячейках > 10 => воспользуемся критерием Хи-квадрат

```python
scipy.stats.chi2_contingency(resualt, correction=False)
```

p-val < 0.05 => мы отвергаем гипотезу о независимости переменных (Gender и JobPref зависимы)

Рассмортим значения переменной Income

```python
survey_df.describe()
```

Будем считать что доход до 20000 - низкий, от 20000 до 60000 - средний, > 60000 - высокий 

```python
survey_df["income_level"] = pd.cut( survey_df["Income"], \
                                   bins=[0, 20000, 60000, float("inf")], \
                                   labels=["низкий", "средний", "высокий"])
survey_df
```

Исследуем связи между Gender и Income_level с помощью таблиц сопряженности 

```python
survey_df.Gender.unique()
```

```python
survey_df.income_level.unique()
```

```python
df_4 = survey_df[["Gender", "income_level"]].dropna()
df_4
```

```python
observed_4 = pd.crosstab(df_4.Gender, df_4.income_level )
observed_4
```

```python
expected_4 = scipy.stats.contingency.expected_freq(observed_4)
expected_4
```

В таблице нет значений < 10 => воспользуемся критерием Хи-квадрат 

```python
scipy.stats.chi2_contingency(observed_4, correction=False)
```

p-val < 0.05 => мы отвергаем гипотезу о независимости переменных (Gender и JobWherePref зависимы)
